```python
# Creaing tables
import sqlite3

# Create a SQLite database connection
conn = sqlite3.connect('hoteldetails.db')

# Create a cursor object
cursor = conn.cursor()

# Create and insert data into Data1 table
cursor.execute('''CREATE TABLE IF NOT EXISTS Data1 (SNo INTEGER, Hotelname VARCHAR, City VARCHAR, Location VARCHAR,
Numberofreviews INTEGER,
Userrating INTEGER,
Starrating INTEGER,
Reviewscore VARCHAR
)''')

# Create and insert data into Data2 table
cursor.execute('''CREATE TABLE IF NOT EXISTS Data2 ( SNo INTEGER, Hotelname VARCHAR, Freewifi VARCHAR, Familyrooms VARCHAR, 
NonSmokingrooms VARCHAR,
Restaurant VARCHAR,
Bar VARCHAR,
Heating VARCHAR,
Lift VARCHAR,
BreakfastCuisine1 VARCHAR,
BreakfastCuisine2 VARCHAR,
BreakfastCuisine3 VARCHAR
)''')

# Create and insert data into Data3 table
cursor.execute('''CREATE TABLE IF NOT EXISTS Data3 ( SNo INTEGER, Hotelname VARCHAR, Roomtype VARCHAR, Doublebed VARCHAR,
Singlebed VARCHAR,
Price INTEGER
)''')

# Commit the changes and close the connection
conn.commit()
```

```python
import csv

# Define the CSV file paths
data1_csv = "Data1.csv"
data2_csv = "Data2.csv"
data3_csv = "Data3.csv"

# Function to insert data from a CSV file into a table
def insert_data_from_csv(csv_file, table_name, encoding='utf-8'):
    with open(csv_file, 'r', newline='', encoding=encoding) as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip the header row

        for row in reader:
            # Insert data from the CSV into the SQLite table
            cursor.execute(f"INSERT INTO {table_name} VALUES ({', '.join(['?'] * len(row))})", row)

# Insert data from CSV files into the corresponding tables
insert_data_from_csv(data1_csv, 'Data1')
insert_data_from_csv(data2_csv, 'Data2')
insert_data_from_csv(data3_csv, 'Data3')

# Commit the changes and close the connection
conn.commit()
```
